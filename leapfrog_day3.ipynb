# Handling Missing **Data**

```python
# Check missing values
print(df.isnull().sum())

# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop 'Cabin' column due to too many missing values
df.drop(columns=['Cabin'], inplace=True)
```

# Creating New Columns

```python
# Create Age Group column
df['AgeGroup'] = df['Age'].apply(lambda age: 'Child' if age < 18 else 'Adult')

# Create IsAlone column
df['IsAlone'] = (df['SibSp'] + df['Parch'] == 0)
```

# GroupBy & Aggregation

```python
# Average fare by class
print(df.groupby('Pclass')['Fare'].mean())

# Survival rate by gender
print(df.groupby('Sex')['Survived'].mean())

# Mean age by gender and class
print(df.groupby(['Sex', 'Pclass'])['Age'].mean())
```

#Correlation Analysis

```python
# Numeric correlation matrix
print(df.corr(numeric_only=True))

# Correlation between Fare and Survival
print("Fare vs Survived Corr:", df['Fare'].corr(df['Survived']))
```

#  Plot Titanic Dataset Example

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")


```

# Histogram – Age Distribution

```python
plt.hist(df['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.grid(True)
plt.show()
```

# Bar Chart – Survival Count

```python
survived_count = df['Survived'].value_counts()
plt.bar(['Not Survived', 'Survived'], survived_count, color=['salmon', 'seagreen'])
plt.title("Survival Count")
plt.ylabel("Number of Passengers")
plt.show()
```

# Bar Chart – Survival by Gender

```python
gender_survival = df.groupby('Sex')['Survived'].sum()
plt.bar(gender_survival.index, gender_survival, color=['pink', 'lightblue'])
plt.title("Survived by Gender")
plt.ylabel("Number of Survivors")
plt.show()
```

# Boxplot – Fare by Class

```python
df.boxplot(column='Fare', by='Pclass', grid=False)
plt.title("Fare Distribution by Passenger Class")
plt.suptitle("")  # Remove default title
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

#  Scatter Plot – Age vs Fare

```python
plt.scatter(df['Age'], df['Fare'], alpha=0.5, c=df['Survived'], cmap='bwr')
plt.title("Fare vs Age (Color = Survived)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.colorbar(label='Survived (0=No, 1=Yes)')
plt.show()
```

###  Plot Types – What They Show & When to Use

|  **Chart Type** |  **What It Shows**                                                                | 🛠️ **When to Use**                                                                                                            |
|------------------|--------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------------------------------|
| **Histogram**     | Frequency distribution of a **single numerical variable** (e.g., Age or Fare)       | ✔️ Use to **understand distribution**, detect **skewness**, **outliers**, or **common ranges** in numerical data.             |
| **Bar Chart**     | Comparison of **categories** (e.g., survival count, gender, passenger class)        | ✔️ Use to **compare quantities** across **discrete categories** (e.g., how many survived in each gender or class).            |
| **Boxplot**       | Spread, median, quartiles, and outliers of **numerical data grouped by categories** | ✔️ Use to **compare distributions** and **spot outliers** (e.g., Fare by Pclass, Age by Gender).                              |
| **Scatter Plot**  | Relationship between **two numerical variables** (e.g., Age vs Fare)                | ✔️ Use to explore **correlations**, **clusters**, or **trends** in continuous data (great before regression).                 |
| **Subplots**      | Multiple plots shown side-by-side or in a grid                                      | ✔️ Use to **compare multiple visualizations together** (e.g., Age distribution + Gender breakdown).                           |


### Titanic Plot Types Summary

| Chart Type   | Titanic Example                           | Purpose                                 |

| **Histogram**    | Distribution of passenger **Age**         | Understand age groups on the ship       |

###  Choosing the Right Plot – Quick Guide

| What You Want to Know               | Use This Plot |
|------------------------------------|---------------|
| "How is this variable distributed?" | **Histogram** |
| "Which category has more or less?"  | **Bar Chart** |
| "Are there any outliers in groups?" | **Boxplot**   |
| "Do two values have a pattern?"     | **Scatter Plot** |
| "Can I see many things together?"   | **Subplots**  |


 Pro Tips

    Use histograms before normalization or binning.

    Use bar charts after grouping or counting (groupby() + count() or sum()).

    Use boxplots to clean or understand spread before modeling.

    Use scatter plots when testing for linear or non-linear relationships.

    Use subplots in dashboards or reports to tell a story with visuals.
