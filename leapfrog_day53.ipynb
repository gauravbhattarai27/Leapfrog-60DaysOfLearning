```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Get dataset names
sns.get_dataset_names()
```

```python
# Barplot
df = sns.load_dataset('diamonds')
df.head()
```

```python
import pandas as pd
df2 = pd.DataFrame({
    'Items': ['Laptop', 'Watch', 'Phone', 'Other'],
    'Sales': [400,600,200,700]
})
```

```python
df2
```

```python
# .bar()
plt.bar(df2.Items, df2.Sales)
plt.title('Sales of Items')
plt.xlabel('Items')
plt.ylabel('Sales')
plt.show()
```

```python
# Common Matplotlib Functions
plt.bar()
plt.scatter()
plt.hist()
plt.plot()
plt.boxplot()
```

```python
# Using Seaborn
sns.barplot(data = df2, x = 'Sales', y= 'Items') # Switch x and y to switch between horizontal and vertical
# plt.xlabel('Commodities')
plt.show()
```

```python
# value_counts()
count = df.cut.value_counts()
count
```

```python
plt.bar(count.index, count.values)
plt.title('Diamond categories by Cut')
plt.xlabel('Cut')
plt.ylabel('Count')
plt.show()
```

```python
# Using Seaborn
sns.countplot(data = df, x = 'cut')
plt.show()
```
