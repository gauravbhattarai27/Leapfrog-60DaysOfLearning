```python
import pandas as pd

df = pd.read_csv('/content/wine.csv')
df.head()
```

```python
# Basic structure
df.info()

# Summary statistics
df.describe()

# Check for nulls
df.isnull().sum()
```

```python
# Univariate Statistics
# Example on 'alcohol' column
df['Alcohol'].mean()
df['Alcohol'].median()
df['Alcohol'].mode()[0]
```

```python
df['Alcohol'].var()
df['Alcohol'].std()
```

```python
df['Alcohol'].skew()
df['Alcohol'].kurt()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
sns.histplot(df['Alcohol'], kde=True)
plt.title('Alcohol Distribution')
plt.show()

# Boxplot
sns.boxplot(x=df['Alcohol'])
plt.title('Alcohol Boxplot')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()

# Covariance
df.cov()
```

```python
prob = (df['Alcohol'] > 13.0).mean()
print(f"Probability(alcohol > 13.0) = {prob:.2f}")
```
